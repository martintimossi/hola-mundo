```python
## condicionales para tomar decisiones

def revision_patentes (dominio):
  """
  esta funcion evalua el ingreso de patentes revisando mediante condicionales
  si los primeros 3 caracteres son letras y los segundos 3 son numeros

  """

  numero = dominio[-3:]
  letras = dominio[:3]

  print("letras ingresadas: " + letras + "\n")
  print("parte numerica ingresada: " + numero + "\n")

  if letras.isalpha():
    if numero.isdigit():
      print("la patente " + dominio + " es correcta")
    else:
      print("los ultimos 3 deben ser numeros")
  else:
    print("las primeras 3 deben ser letras")


revision_patentes("MQR836")
```

```python
# Las funciones que devuelven múltiples valores los devuelven en una tupla.
def  to_dollars_cents ( precio ):
    '''
    Divide el precio (flotante) en dólares y centavos.
    '''
    dólares  =  int ( precio  //  1 )
    centavos  =  round ( precio  %  1  *  100 )

    return  dólares ,  centavos

to_dollars_cents(12.34)
```

```python
## desempaquetar (pasa de tupla a enteros por separado)

dolares, centavos = to_dollars_cents(89.78)
print(dolares)
print(centavos)

type(centavos)
```

### Agrupamiento y Agregacion

```python
import numpy as np
import pandas as pd

# nuevo marco de datos con mas datos de planetas

data = {'planet': ['Mercury', 'Venus', 'Earth', 'Mars',
                   'Jupiter', 'Saturn', 'Uranus', 'Neptune'],
        'radius_km': [2440, 6052, 6371, 3390, 69911, 58232,
                     25362, 24622],
        'moons': [0, 0, 1, 2, 80, 83, 27, 14],
        'type': ['terrestrial', 'terrestrial', 'terrestrial', 'terrestrial',
                 'gas giant', 'gas giant', 'ice giant', 'ice giant'],
        'rings': ['no', 'no', 'no', 'no', 'yes', 'yes', 'yes','yes'],
        'mean_temp_c': [167, 464, 15, -65, -110, -140, -195, -200],
        'magnetic_field': ['yes', 'no', 'yes', 'no', 'yes', 'yes', 'yes', 'yes']
        }

planetas = pd.DataFrame(data)

planetas
```

```python
# agrupar por tipo

# agrupa por tipo y con "sum()" le decimos que sume todos los valores agrupados

planetas.groupby(['type']).sum()
```

```python
# agrupamiento con vizualizacion de solo una columna

planetas.groupby(['type']).sum()[['moons']]
```

```python
# agrupar con dos criterios (tipo y campo magnetico) y calcular el promedio de lunas, radio km y temp de gada grupo

planetas.groupby(['type', 'magnetic_field'])[['moons', 'radius_km', 'mean_temp_c']].mean()
```

```python
# usar la funcion agregacion "agg()" para agregar mas resultados a las variables agrupadas


planetas.groupby(['type', 'magnetic_field'])[['radius_km', 'moons']].agg(['mean', 'max'])

```

```python
# funcion que calcula el 90 porciento de un nro

def percentile_90(x):
    return x.quantile(0.9)
```

```python
# usar funciones creadas en agrupamiento y agregacion

planetas.groupby(['type', 'magnetic_field'])[['radius_km', 'moons']].agg(['max', percentile_90])
```

### Fusion y Unnion de Datos

```python
import numpy as np
import pandas as pd

# nuevo marco de datos planetas
data = {'planet': ['Mercury', 'Venus', 'Earth', 'Mars'],
        'radius_km': [2440, 6052, 6371, 3390],
        'moons': [0, 0, 1, 2],
        }
# convertir diccionario en marco de datos.
df1 = pd.DataFrame(data)
df1
```

```python
# marco de datos planetas 2

data = {'planet': ['Jupiter', 'Saturn', 'Uranus', 'Neptune'],
        'radius_km': [69911, 58232, 25362, 24622],
        'moons': [80, 83, 27, 14],
        }
# convertir diccionario en marco de datos.
df2 = pd.DataFrame(data)
df2
```

```python
# concatenar un marco de datos encima de otro (vertical)

df3 = pd.concat([df1,df2], axis=0)

df3
```

```python
# reiniciar numeracion para seguir de manera continua

df3 = df3.reset_index(drop=True)
df3
```

```python
# marco de datos con mismas filas y nuevas columnas

data = {'planet': ['Earth', 'Mars','Jupiter', 'Saturn', 'Uranus',
                   'Neptune', 'Janssen', 'Tadmor'],
        'type': ['terrestrial', 'terrestrial','gas giant', 'gas giant',
                 'ice giant', 'ice giant', 'super earth','gas giant'],
        'rings': ['no', 'no', 'yes', 'yes', 'yes','yes', 'no', None],
        'mean_temp_c': [15, -65, -110, -140, -195, -200, None, None],
        'magnetic_field': ['yes', 'no', 'yes', 'yes', 'yes', 'yes', None, None],
        'life': [1, 0, 0, 0, 0, 0, 1, 1]
        }
df4 = pd.DataFrame(data)

df4
```

```python
# unir marcos de datos a partir de la variable "planet"

inner = pd.merge(df3, df4, on='planet', how='inner') # "INNER" SOLO CONSERVA LOS DATOS QUE SE ENCUENTRAN EN AMBOS DF
inner
```

```python
# "OUTER" CONSERVA TODOS LOS DATOS COMPPLETANDO CON "NaN"

outer = pd.merge(df3, df4, on='planet', how='outer')
outer
```

```python
# "LEFT" CONSERVA TODOS LOS DATOS QUE SE ENCUENTRAN EN DF IZQUIERDO

left = pd.merge(df3, df4, on='planet', how='left')
left
```

```python
# "LEFT" CONSERVA TODOS LOS DATOS QUE SE ENCUENTRAN EN DF DERECHO

right = pd.merge(df3, df4, on='planet', how='right')
right
```

## Muestreo: muestreo de la población, teorema limite central


**El proceso de muestreo**

En primer lugar, repasemos las principales etapas del proceso de muestreo:

1. Identificar la población objetivo

2. Seleccionar el marco de muestreo

3. Elegir el método de muestreo

4. Determinar el tamaño de la muestra

5. Recoger los datos de la muestra

**Paso 1: Identificar la población objetivo**

La población objetivo es el conjunto completo de elementos sobre los que le interesa saber más.
La población objetivo podria ser los 10.000 clientes que compraron el último modelo de frigorífico de la empresa.

**Paso 2: Seleccionar el marco de muestreo**
Un marco de muestreo es una lista de todos los individuos o elementos de la población objetivo.

La diferencia entre una población objetivo y un marco de muestreo es que la población es general y el marco es específico. **Métodos de Muestreo probabilístico**

Existen cuatro métodos diferentes de muestreo probabilístico:

1. *Muestreo aleatorio simple:* Cada miembro de una población se selecciona aleatoriamente y tiene la misma probabilidad de ser elegido.

2. *Muestreo aleatorio estratificado:*  Se divide la población en grupos y se seleccionan al azar algunos miembros de cada grupo para que formen parte de la muestra. Ej: algunos empleados de diferentes sucursales de una cadena de heladerias.

3. *Muestreo aleatorio por conglomerados:* Se divide una población en conglomerados, se seleccionan aleatoriamente determinados conglomerados y se incluyen en la muestra todos los miembros de los conglomerados elegidos. Ej: El total de empleados de algunas heladerias especificas de la cadena.

4. *Muestreo aleatorio sistemático:* Se coloca a cada miembro de una población en una secuencia ordenada. A continuación, se elige un punto de partida aleatorio en la secuencia y se seleccionan miembros para la muestra a intervalos regulares. Ej: Pone los nombres de los estudiantes en orden alfabético y elegiría al azar un punto de partida.

## Testeo de Hipótesis


La prueba de hipótesis utiliza datos de muestra para evaluar una hipótesis sobre un parámetro de población.

Ejemplo: El cambio de color de una pagina web incrementa la permanencia de los usuarios en linea.

Pasos para realizar una prueba de hipótesis:

1. Enuncie la hipótesis nula y la hipótesis alternativa.

2. Elija un nivel de significación.

3. Determinar el Valor P.

4. Rechazar o no rechazar la hipótesis nula.


**Hipótesis nula** es una afirmación que se asume como cierta a menos que haya pruebas convincentes de lo contrario. La hipótesis nula suele suponer que no hay ningún efecto en la población y que los datos observados se producen por casualidad.

**hipótesis alternativa** es una afirmación que contradice la hipótesis nula y sólo se acepta como cierta si existen pruebas convincentes de lo contrario. La hipótesis alternativa suele suponer que existe un efecto en la población y que los datos observados no se producen por casualidad.


**Nivel de Significancia**: El valor estándar suele ser 0.05 (5%).Esto significa que estamos dispuestos a aceptar un 5% de probabilidad de equivocarnos y decir que hay un efecto cuando en realidad no lo hay (esto se llama Error Tipo I).

**Valor p**: Es la probabilidad de haber obtenido los resultados observados

***Resultado final:***

*   Si el P-valor es menor que $\alpha$ (p < 0.05): Rechazamos la
$H_0$(Hipotesis Nula). Los resultados son "estadísticamente significativos".
*   Si el P-valor es mayor que $\alpha$ (p > 0.05): No podemos rechazar la $H_0$(Hipotesis Nula). No hay evidencia suficiente para asegurar un cambio.
